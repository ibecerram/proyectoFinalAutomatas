### Supermercado
### Un supermercado con N productos tiene que buscarse la forma de recorrerlo.


```python
class Node:
    def __init__(self, hallway, height):
        self.hallway = hallway
        self.height = height
        self.listProduct = []
    
    def addProduct(self, product):
        self.listProduct.append(product)
    
    def __repr__(self):
        return str(self.hallway) + " , " + str(self.height)
        #return "Node"
```

```python
import math
from collections import defaultdict
import networkx as nx
import matplotlib.pyplot as plt

class Graph:
    def __init__(self, hallways, height):
        self.listNodes = []
        self.hallways = hallways
        self.height = height
        self.edges = defaultdict(list)
    
    def setListNodes(self):
        for hallway in range(self.hallways):
            for height in range(self.height):
                #print(hallway,",", height)
                self.listNodes.append(Node(hallway, height))
                
    def addEdge(self, nodeA, nodeB):
        self.edges[nodeA].append([nodeB, 5])
        self.edges[nodeB].append([nodeA, 5])
    
    def setEdges(self):
        x = 0
        for hallway in range(self.hallways):
            
            for height in range(self.height):
                if height < self.height - 1:
                    #print("Posición: ", self.listNodes[x])
                    nodeA = self.listNodes[x]
                    nodeB = self.listNodes[x+1]
                    self.addEdge(nodeA, nodeB)
                else:
                    if hallway < self.height - 1:
                        #print(x)
                        #print(self.listNodes[x])
                        #print(self.listNodes[x+self.hallways])
                        
                        nodeA = self.listNodes[x + self.hallways]
                        nodeB = self.listNodes[x]
                        self.addEdge(nodeA, nodeB)
                x = x+1;
                

                
            if hallway + 1 < self.hallways:
                #print(self.listNodes[x - self.hallways])
                #print(self.listNodes[x])
                
                nodeA = self.listNodes[x - self.hallways]
                nodeB = self.listNodes[x]
                self.addEdge(nodeA, nodeB)
            
    
    def addNode(self, node):
        self.listNodes.append(node)
        
    def showNetworkX(self):
        G = nx.Graph()
        G.add_nodes_from(self.listNodes)
        
        for edge, values in self.edges.items():
            for arista in values:
                G.add_edge(edge, arista[0], weight = arista[1])
        
        pos = nx.spring_layout(G, k=0.99, iterations=500)
        nx.draw(G, pos, with_labels=True)
        edge_labels = dict([((n1, n2), d['weight']) for n1, n2, d in G.edges(data=True)]) #Aquí debe ser Data = True

        nx.draw_networkx_edge_labels(G, pos, font_size=13, edge_labels=edge_labels)
        plt.show()
        
```

```python
class Dijkstra:
    def __init__(self, graph):
        self.graph = graph
        self.listDistances = defaultdict(list)
        self.listPadre = defaultdict(list)
        self.listVistos = defaultdict(list)
        self.stack = []
        self.path = []
        self.exit = False
        
        self.initialize()
    
    def initialize(self):
        for node in self.graph.listNodes:
            self.listDistances[node] = (math.inf)
            self.listPadre[node] = None
            self.listVistos[node] = False
    
    def clear(self):
        self.listDistances = defaultdict(list)
        self.listPadre = defaultdict(list)
        self.listVistos = defaultdict(list)
        self.stack = []
        self.path = []
        self.exit = False
        self.initialize()
    
    def extractMinimum(self):
        minimum = self.stack[0]
        for node in self.stack:
            if node[1] < minimum[1] and self.listVistos[node[0]] is False:
                minimum = node
                
        self.stack.remove(minimum)
        return minimum[0]
    
    def showResults(self):
        #print("Distancias obtenidas: ")
        #print(self.listPadre)
        for distance, value in self.listDistances.items():
            #print(distance)
            print("\nNodo: ", distance, " Distance: [", value, ", Parent: ", self.listPadre[distance], " ]")
        
        self.graph.showNetworkX()
            
    def runAlgorithm(self, nodeOrigin):
        self.listDistances[nodeOrigin] = 0
        self.stack.append((originNode, self.listDistances[originNode]))
        
        while self.stack:
            
            u = self.extractMinimum()
            self.listVistos[u] = True
            
            for edges in self.graph.edges[u]:
                flag = self.listVistos.get(edges[0])
                if flag is False:
                    if self.listDistances[edges[0]] > (self.listDistances[u] + edges[1]):
                        self.listDistances[edges[0]] = self.listDistances[u] + edges[1]
                        self.listPadre[edges[0]] = u
                        self.stack.append((edges[0], self.listDistances[edges[0]]))
        
        self.showResults()
        
    def findShortNode(self, nodeOrigin, listNodes):
        if self.exit == False:
            self.clear()
            self.listDistances[nodeOrigin] = 0
            self.stack.append((originNode, self.listDistances[originNode]))

            while self.stack:

                u = self.extractMinimum()
                self.listVistos[u] = True

                for edges in self.graph.edges[u]:
                    flag = self.listVistos.get(edges[0])
                    if flag is False:
                        if self.listDistances[edges[0]] > (self.listDistances[u] + edges[1]):
                            self.listDistances[edges[0]] = self.listDistances[u] + edges[1]
                            self.listPadre[edges[0]] = u
                            self.stack.append((edges[0], self.listDistances[edges[0]]))
            
            #print("ACTUAL: ", nodeOrigin)
            shortNode = self.getShortNode(listNodes)
            #self.path.append(shortNode)
            #print("MAS CERCA: ", shortNode)
            #print("Nodos restantes: ", listNodes)
            #listNodes.remove(shortNode)
            #print("Despues de borrar nodo: ", listNodes)

            self.checkPath(shortNode, nodeOrigin, listNodes)



            return shortNode
        else:
            print("SALIENDO....");
     
    def checkPath(self, node, origin, listNodes):
        #print("Nodo origen: ", origin)
        #print("Nodo destino: ", node)
        
        while node != origin:
            #print("AQUI: ", self.listPadre[node])
            #print(self.listPadre)
            node = self.listPadre[node]
            self.path.insert(0,node)
        #print(self.path)
        
    def getShortNode(self, listNodes):
        shortNode = listNodes[0]
        minShort = 999
        
        
        for distance, value in self.listDistances.items():
            if distance in listNodes:
                if value < minShort:
                    shortNode = distance
                    minShort = value
            #print("\nNodo: ", distance, " Distance: [", value, ", Parent: ", self.listPadre[distance], " ]")
        
        #print("NODO MÄS CERCA: ", shortNode)
        return shortNode
     
    def findShortestPath(self, origin, listNodes):
        nodos = len(listNodes)
        
        
        while nodos != 0:
            #print(self.listDistances)
            origin = self.findShortNode(origin, listNodes)
            print(origin)
            listNodes.remove(origin)
            nodos = len(listNodes)
            
        print("---------")
        print("Origen Final: ", origin)
        listNodes.append(self.graph.listNodes[24])
        print(listNodes)
        origin = self.findShortNode(origin, listNodes)
        #self.path.append(origin)
        print(self.path)
```

```python
graph = Graph(5,5)
graph.setListNodes()
graph.setEdges()

#Nodo Inicial
originNode = graph.listNodes[0]

#Nodos que visitará
listNodes = []
listNodes.append(graph.listNodes[11])
listNodes.append(graph.listNodes[4])
listNodes.append(graph.listNodes[22])

nodos = len(listNodes)
flag = True
path = []
while flag:
    if nodos != 0:
        dijkstra = Dijkstra(graph)
        originNode = dijkstra.findShortNode(originNode, listNodes)
        listNodes.remove(originNode)
        nodos = len(listNodes)
        path.extend(dijkstra.path)
    else:
        dijkstra = Dijkstra(graph)
        listNodes.append(graph.listNodes[20])
        originNode = dijkstra.findShortNode(originNode, listNodes)
        listNodes.remove(originNode)
        flag = False
        path.extend(dijkstra.path)
        
print("Path Completo: ", path)
```
